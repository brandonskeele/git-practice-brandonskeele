```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url = "https://github.com/esnt/Data/raw/main/CleanData/tornadoes.csv"
df = pd.read_csv(url)
df["lon"] = df["lon"] * -1
```

```python
df.head()
```

```python
px.scatter_geo(df, lon="lon", lat="lat",scope="usa")
```

```python
fig = px.scatter(df, x="lon", y="lat")
fig.update_layout(width=800, height=800)
```

```python
px.scatter(df, x="lon", y="lat", hover_data=["state", "year"], color="month")
```

```python
px.box(df, x="max_width")
```

```python
px.set_mapbox_access_token(open("mapbox_apikey.txt").read())
```

```python
fig = px.scatter_mapbox(df, lon="lon", lat="lat", zoom = 3)
fig.update_layout(width=800, height=600)

# fig.write_html("map.html")
fig.show()
```
